```python
import seaborn as sns
import matplotlib.pyplot as plt

# Get the top 5 'ship-service-level' values
top_values = df['ship-service-level'].value_counts().nlargest(5).index

# Filter the dataframe to include only the top 5 values
df_top5 = df[df['ship-service-level'].isin(top_values)]

# Create the countplot
ax = sns.countplot(data=df_top5, x='ship-service-level', hue='ship-city', dodge=True)

# Label the bars
for container in ax.containers:
    ax.bar_label(container)

# Show the plot
plt.show()
```
